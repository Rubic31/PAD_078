```python
import numpy as np
import plotly.graph_objects as go
import pandas as pd
import scipy.stats as stats
#import utils

def student_normal(x):
    mean, std = 175, 4
    exponent = ((x-mean)/std)**2
    exponent *= -0.5
    output = np.exp(exponent)/(std*np.sqrt(2*np.pi))
    return output
```

```python
offset_data = pd.read_csv("/content/drive/MyDrive/PAD/LAB07/wyniki.csv")
```

```python
offset_data
```

```python
female_data = offset_data[offset_data['plec'] == 'F']
male_data = offset_data[offset_data['plec'] == 'M']
```

```python
male_data
```

```python
# Computing test statistic

male_results = male_data['ocena_koncowa']
female_results = female_data['ocena_koncowa']

sample_std_male = np.std(male_results)
sample_mean_male = np.mean(male_results)
n_male = male_results.shape[0]

sample_std_female = np.std(female_results)
sample_mean_female = np.mean(female_results)
n_female = female_results.shape[0]



t_statistic = (sample_mean_male - sample_mean_female)/(np.sqrt(np.square(sample_std_male)/n_male + np.square(sample_std_female)/n_female))
print("t-statistic:", t_statistic)
```

```python
df = offset_data.shape[0]-1
p = stats.t.sf(abs(t_statistic), df=df)
```

```python
p
```

```python
# p value jest mniejsze niż significance level 0.05 a więc hipoteze zerową można odrzucić
# hipoteza alternatywna jest prawidłowa
```

```python
value_data = pd.read_csv('/content/drive/MyDrive/PAD/LAB07/ZyskiFirmyX.csv')
```

```python
value_data
```

```python
value_data['Rok'].describe()
```

```python
import plotly.express as px

px.scatter(value_data, "Rok", "Zysk", title="Year vs Zysk")
```

```python
import statsmodels.formula.api as smf
```

```python
import statsmodels.formula.api as smf

model = smf.ols(formula="Zysk ~ Rok", data=value_data).fit()
print(model.summary())
```

```python
import plotly.graph_objects as go

value_data["fitted"] = model.fittedvalues

fig = go.Figure()
fig.add_trace(go.Scatter(
    x=value_data["Rok"], y=value_data["Zysk"], name="Year vs Profit", mode="markers"))
fig.add_trace(go.Scatter(
    x=value_data["Rok"], y=value_data["fitted"], name="Fitted Regression Line"))
fig.update_layout(title="Regression line of Year vs profit", xaxis_title="Year",
    yaxis_title="Profit")
fig.show()
```

```python
# trend jest raczej stały
value_data["YearAdj"] = value_data["Rok"] - 1821
model = smf.ols(formula="Zysk ~ YearAdj", data=value_data).fit()
print(model.summary())
```

```python
import plotly.graph_objects as go
```

```python
value_data["fitted"] = model.fittedvalues
```

```python
fig = go.Figure()
# Note that we still plot against the original Year variable
fig.add_trace(go.Scatter(
    x=value_data["Rok"], y=value_data["Zysk"], name="Year vs Profit", mode="markers"))
fig.add_trace(go.Scatter(
    x=value_data["Rok"], y=value_data["fitted"], name="Fitted Regression Line"))
fig.update_layout(title="Regression line of Year vs Profit", xaxis_title="Year",
    yaxis_title="Profit")
fig.show()
```
